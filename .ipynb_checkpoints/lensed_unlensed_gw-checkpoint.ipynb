```python
# computation of  binary masses for given chirp mass and q

chirp_mass = np.arange(10,70,10)
mass_ratio = np.arange(0.1, 1.1, 0.1)

# check arrays
#print("Chirp Masses:", chirp_mass)
#print('Mass Ratio, q:', mass_ratio)

# dataframe for chirp mass, q, m1, m2
# mainly for determining binary masses that will give the desired chirp mass and mass ratios 
# binary masses m1 and m2 will be used to generate an unlensed GW signal

chirp_mass_col = chirp_mass.tolist()*len(mass_ratio)
chirp_mass_col.sort()

mass_ratio_col = mass_ratio.tolist()*len(np.unique(chirp_mass))
        
col_dict = {'chirp_mass':chirp_mass_col,
           'q': mass_ratio_col,
           'm1':'',
           'm2':'',
           #'chirp_mass_check':'',
           #'q_check':''
           'Mf':'',
           'f0':'',
           'T0':'',
           'td':''}

cbc_params = pd.DataFrame(col_dict)

for n in np.arange(0, len(cbc_params['chirp_mass'])):
    cbc_params.at[n,'m1'] = cbc_params['chirp_mass'][n]*(((cbc_params['q'][n])**3)/(1+cbc_params['q'][n]))**(-1/5)
    cbc_params.at[n,'m2'] = cbc_params['q'][n]*cbc_params['m1'][n]
    #cbc_params.at[n,'chirp_mass_check'] = (((cbc_params['m1'][n]*cbc_params['m2'][n])**3)/(cbc_params['m1'][n]+cbc_params['m2'][n]))**(1/5)
    #cbc_params.at[n,'q_check'] = cbc_params['m2'][n]/cbc_params['m1'][n]
    
display(cbc_params)

cbc_params.to_csv("/Users/genebelinario/Documents/dnc_research/thesis/master_files/cbc_params.csv")
```

```python
# ini files for each binary mass inputs m1 and m2
# for generating unlensed GW
# /Users/genebelinario/Documents/dnc_research/thesis/master_files/cbc_mass_inputs_ini

from lensinggw.waveform.waveform import gw_signal

# SNR_list contains dictionary of the signal-to-noise ratio 
SNR_list = []

# sH1_list contains list of strain dictionaries, (H1 : strain, strain, strain, ...)
sH1_list = []

# freqs contains frequency values
freqs_list = []

for n in np.arange(0, len(cbc_params.index)):
    
    ini_file = '/Users/genebelinario/Documents/dnc_research/thesis/master_files//cbc_mass_inputs_ini/waveform_config.ini'
    
    config = configparser.ConfigParser()
    config.optionxform = str
    config.read(ini_file)

    config.set('injection parameters', 'm1', str(cbc_params['m1'][n]))
    config.set('injection parameters', 'm2', str(cbc_params['m2'][n]))

    with open(ini_file, 'w') as configfile:
        config.write(configfile)
    
    # instantiate the waveform model
    waveform_model = gw_signal(ini_file)
    
    # compute the unlensed waveform polarizations, strains in the requested detectors and their frequencies
    freqs, hp_tilde, hc_tilde, strain_dict = waveform_model.unlensed_gw()
    
    # and their signal-to-noise-ratios
    SNR_dict = waveform_model.unlensed_snr()
    
    # access an unlensed strain
    sH1 = strain_dict['H1']
    
    SNR_list.append(SNR_dict)
    sH1_list.append(strain_dict)
    freqs_list.append(freqs)
    
```

```python
# we only take the case source position RA,Dec = (0,0.5)thetaE

res = pd.read_csv("/Users/genebelinario/Documents/dnc_research/thesis/master_files/output.csv", 
                          index_col=[0])

for n in np.arange(0,18,3):
    res.drop(n, axis=0, inplace=True)
    res.drop(n+2, axis=0, inplace=True)
    
res.reset_index(inplace=True)
res.drop('index', axis=1, inplace=True)

display(res)
```

```python
# SNR_list contains dictionary of the lensed signal-to-noise ratio 
SNR_list_lensed = []

# sH1_list contains list of lensed strain dictionaries, (H1 : strain, strain, strain, ...)
sH1_list_lensed = []

# freqs contains frequency values
freqs_list_lensed = []

# source position, in radians
beta0 = res['source_ra'][3]*res['thetaE'][3]
beta1 = res['source_dec'][3]*res['thetaE'][3]

# redshifts
zL = res['zL'][3] 
zS = res['zS'][3]

# lens system
lens_model_list     = ['SIE'] 
kwargs_SIE         = {"center_x":0.0, 'center_y':0.0, 'theta_E':res['thetaE'][3], 'e1':-0.081, 'e2':0}
kwargs_lens_list    = [kwargs_SIE]  

# binary point mass images, in radians
ra  = np.array([x for x in res.loc[3,'img0_ra':'img3_ra'] if np.isnan(x) == False])
dec = np.array([x for x in res.loc[3,'img0_dec':'img3_dec'] if np.isnan(x) == False])

# lensed waveforms #
from lensinggw.waveform.waveform import gw_signal

for n in np.arange(0, len(cbc_params.index)):
    
    ini_file = '/Users/genebelinario/Documents/dnc_research/thesis/master_files//cbc_mass_inputs_ini/waveform_config.ini'
    
    config = configparser.ConfigParser()
    config.optionxform = str
    config.read(ini_file)

    config.set('injection parameters', 'm1', str(cbc_params['m1'][n]))
    config.set('injection parameters', 'm2', str(cbc_params['m2'][n]))

    with open(ini_file, 'w') as configfile:
        config.write(configfile)

    # instantiate the waveform model
    waveform_model = gw_signal(ini_file)

    # compute the lensed waveform polarizations, strains in the requested detectors and their frequencies
    freqs_lensed, hp_tilde_lensed, hc_tilde_lensed, lensed_strain_dict = waveform_model.lensed_gw(ra,dec,
                                                                                              beta0,beta1,
                                                                                              zL, zS,
                                                                                              lens_model_list,
                                                                                              kwargs_lens_list)

    # and their signal-to-noise-ratios
    lensed_SNR_dict = waveform_model.lensed_snr(ra,dec,
                                            beta0,beta1,
                                            zL, zS,
                                            lens_model_list,
                                            kwargs_lens_list)

    # access a lensed strain
    lensed_sH1 = lensed_strain_dict['H1']
    
    SNR_list_lensed.append(lensed_SNR_dict)
    sH1_list_lensed.append(lensed_strain_dict)
    freqs_list_lensed.append(freqs_lensed)
```

```python
# try accessing a signal from the sH1_list_lensed
# note that this returns h(f)
print(sH1_list_lensed[0]['H1'])
print(freqs_list_lensed[0])
print(len(sH1_list_lensed))
```

```python
# mismatch computation

# frequency values
freqs = freqs_list[0]

# let af be the unlensed signal
af = sH1_list[0]['H1']
af_interp_func = interp1d(freqs,af)

# let bf be the lensed signal
bf = sH1_list_lensed[0]['H1']
bf_interp_func = interp1d(freqs,bf)

# interpolate PSD 
psd_file = '/Users/genebelinario/Documents/dnc_research/thesis/master_files/cbc_mass_inputs_ini/LIGO-P1200087-v18-aLIGO_DESIGN_psd.dat'
psd = np.loadtxt(psd_file, unpack=True)

psd_freqs = psd[0]
psd_strain = psd[1]

# interpolate PSD data
psd_interp_func = interp1d(psd_freqs, psd_strain)

# compute mismatch for case M_c =10, q=0.1

psd_interp = psd_interp_func(freqs)

len(psd_interp)
```

```python
# np trapezoidal
func = af*np.conjugate(bf)/psd_interp
result_trap = np.trapz(func, freqs)
print(4*np.real(result_trap))
```

```python
# via scipy
def func_sp(f):
    return af_interp_func(f)*np.conjugate(bf_interp_func(f))/psd_interp_func(f)

result_sp = integrate.quad(func_sp, np.min(freqs), np.max(freqs))
print(4*result_sp[0])
```
